```python
from math import inf
from genres import categorize_genres

info_not_found = pd.DataFrame(not_found_infos)

def parse_date(date):
    try:
        return pd.to_datetime(date).date()
    except:
        return np.nan
    
def parse_runtime(runtime):
    try:
        return int(runtime.split(" ")[0])
    except:
        return np.nan
    
def parse_box_office(bo):
    try:
        return int(bo.replace("$", "").replace(",", ""))
    except:
        return np.nan
    
def parse_countries_genre_language(cgl):
    try:
        return cgl.split(", ")
    except:
        return np.nan
    
def parse_genre(genre):
    try:
        return categorize_genres(genre)
    except:
        return np.nan
    
def parse_review_score(review_score):
    try:
        reviews = [review['Value'] for review in review_score]
        reviews_numbers = []
        for review in reviews:
            if contains(review, "%"):
                reviews_numbers.append(float(review.replace("%", "")) / 100)
            else:
                parts = review.split("/")
                if len(parts) == 2:
                    reviews_numbers.append(float(parts[0]) / float(parts[1]))
                else:
                    reviews_numbers.append(float(parts[0]) / 10)
        return np.mean(reviews_numbers)
    except:
        return np.nan
    
    
def get_awards_received(title):
    return df_oscar[df_oscar['Movie name'] == title].iloc[0]['Awards Won']

def get_awards_nominated(title):
    return df_oscar[df_oscar['Movie name'] == title].iloc[0]['Nominations']
    
info_not_found['release_date'] = info_not_found['release_date'].apply(parse_date)
info_not_found['runtime'] = info_not_found['runtime'].apply(func=parse_runtime)
info_not_found['box_office'] = info_not_found['box_office'].apply(parse_box_office)
info_not_found['countries'] = info_not_found['countries'].apply(parse_countries_genre_language)
info_not_found['languages'] = info_not_found['languages'].apply(parse_countries_genre_language)
info_not_found['genres'] = info_not_found['genre'].apply(parse_countries_genre_language)




info_not_found.drop(columns=['genre'], inplace=True)
info_not_found['reviewScores'] = info_not_found['reviewScore'].apply(parse_review_score)
info_not_found.drop(columns=['reviewScore'], inplace=True)
info_not_found['nbOscarReceived'] = info_not_found['title'].apply(get_awards_received)
info_not_found['nbOscarNominated'] = info_not_found['title'].apply(get_awards_nominated)
info_not_found['wikipedia_id'] = np.nan

```
